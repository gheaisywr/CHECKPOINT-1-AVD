## **Data Preparation**

### **Missing Values**

1. Jumlah_Siswa_Penerima\
pada kolom ini ada beberapa data yang hilang sekitar 36% (69 dari 190 data), kemudian karena 36% itu <=70% berarti data nya ini bisa dihapus atau diimputasi. Disini saya imputasi dengan mengisi data yang kosong menggunakan median, kenapa median? karena kalo konteks datanya adalah jumlah siswa penerima MBG itu sering terdapat perbedaan yang sangat ekstrem antara sekolah di kota besar (yang jumlah siswanya sangat banyak) dengan sekolah di pelosok atau desa terpencil (yang jumlah siswanya sedikit). Jadi nilai lebih stabil kalo pakai median.

```python
df['jumlah_siswa_penerima'] = df['jumlah_siswa_penerima'].fillna(df['jumlah_siswa_penerima'].dropna().median())
```

2. Kalori Per Porsi\
pada kolom ini ada beberapa data yang hilang sekitar 12% (23 dari 190 data), kemudian karena 12% itu <=70% berarti datanya itu bisa di hapus atau diimputasi. Disini saya akan imputasi data dengan dengan mengisi data yang kosong menggunakan mean. kenapa mean? karena datanya tidak memiliki data yang ekstrem banget nilainya sehingga memakai nilai rata-rata nya.

```python
df['kalori_per_porsi'] = df['kalori_per_porsi'].fillna(df['kalori_per_porsi'].dropna().mean())
```

### **Outliers**

```python
Q1 = df["berat_porsi_gram"].quantile(0.25)
Q3 = df["berat_porsi_gram"].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df["berat_porsi_gram"] = df["berat_porsi_gram"].clip(lower=lower_bound, upper=upper_bound)
```

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'berat_per_porsi': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('berat_per_porsi', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('berat_porsi_porsi', axis=1)

# Tampilkan dataframe
display(results_df)
```
